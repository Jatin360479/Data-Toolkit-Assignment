### Question 3. Explain the following terms:
### -The difference in npYarray, npYasarray and npYasanyarrayX
### -The difference between Deep copy and shallow copy.

```python
"""
np.array: Always creates a copy of the input data,
          even if the input is already a NumPy array.
          It ensures that you have a distinct array object.

np.asarray: Attempts to convert the input into a NumPy array
             without creating a copy if the input is already
             an array or array-like (e.g., lists, tuples).
             It's faster and more efficient when you want to
             ensure you have an array without needing a
             separate copy.

np.asanyarray: It is similar to np.asarray, but it does not
                convert the input if it's already a NumPy array.
                It will simply return the original array.
"""

# Example to demonstrate the difference:
import numpy as np

list_data = [1, 2, 3]
array_data = np.array([4, 5, 6])

new_array1 = np.array(list_data)  # Creates a new array
new_array2 = np.asarray(list_data)  # Creates a new array
new_array3 = np.asarray(array_data)  # Returns the existing array
new_array4 = np.asanyarray(array_data)  # Returns the existing array


print(new_array1, new_array2, new_array3, new_array4)
print(new_array3 is array_data)  # True (No new copy was made)
print(new_array4 is array_data)  # True


# ### Explaining the difference between Deep Copy and Shallow Copy


"""
Deep Copy:

- Creates a completely independent copy of the original array.
- Modifications to the deep copy do not affect the original array,
  and vice-versa.
- This involves copying all the elements of the original array into
  a new location in memory.

Shallow Copy:

- Creates a new array object, but the elements of the new array
  are just references (pointers) to the elements of the original array.
- Modifying elements in the shallow copy might affect the original
  array, and vice-versa, because both are still pointing to the
  same underlying data.
- It's quicker and consumes less memory than a deep copy.
"""

# Example:

original_array = np.array([1, 2, 3])

# Deep Copy:
deep_copy_array = original_array.copy()
deep_copy_array[0] = 100
print(f"Original: {original_array}")
print(f"Deep Copy: {deep_copy_array}")  # Modifying deep_copy doesn't change original_array.

# Shallow Copy (View)
shallow_copy_array = original_array.view()
shallow_copy_array[0] = 100

print(f"Original: {original_array}")  # Modifying shallow_copy affects original_array.
print(f"Shallow Copy: {shallow_copy_array}")

```

### Question 14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
### Perform the following tasks using Matplotlib:
### a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

### b) Label the x-axis as 'Date'.

### c) Set the title of the plot as 'Temperature and Humidity over time.

```python
# Create a time-series dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dates = pd.date_range(start='2023-10-26', periods=30)
temperatures = np.random.randint(15, 30, size=30)
humidities = np.random.randint(40, 80, size=30)

df = pd.DataFrame({'Date': dates, 'Temperature': temperatures, 'Humidity': humidities})

# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Create a second y-axis

color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# c) Set the title of the plot
plt.title('Temperature and Humidity over time')

fig.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

### Question 15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
### a) Plot a histogram of the data with 30 bins.

### b) Overlay a line plot representing the normal distribution's probability density function (PDF).

### c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

### d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# a) Plot a histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', label='Histogram')

# b) Overlay a line plot representing the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, loc=0, scale=1)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# c) Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title
plt.title('Histogram with PDF Overlay')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

### Question 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import seaborn as sns

# Create two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
quadrant = np.where(x >= 0, np.where(y >= 0, 1, 4), np.where(y >= 0, 2, 3))

# Create a scatter plot using Seaborn
sns.scatterplot(x=x, y=y, hue=quadrant, palette=['red', 'blue', 'green', 'purple'])

# Add a legend
plt.legend(title='Quadrant')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

### Question 18. With Bo+eh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'8

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

# Generate x and y values for the sine wave
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave Function", x_axis_label="x", y_axis_label="sin(x)")

# Add a line chart
p.line(x, y, line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)
```

### Question 19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'

```python
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=len(categories))

# Create a Bokeh figure
# Changed 'plot_height' to 'height'
p = figure(x_range=categories, height=350, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Create a bar chart with color based on values
p.vbar(x=categories, top=values, width=0.9, color=['red' if val > 10 else 'blue' for val in values])

# Add hover tooltips
from bokeh.models import HoverTool
hover = HoverTool(tooltips=[("Category", "@x"), ("Value", "@top")])
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Show the plot
show(p)
```
